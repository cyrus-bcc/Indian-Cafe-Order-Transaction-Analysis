Sales Analysis

```python
#Import 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Read CSV
df = pd.read_csv('../dataset/preprocessed/cleaned_data.csv')
```

Total Sales

```python
total_sales = df['total'].sum()
print(f"Total Sales: ₹{total_sales:,.2f}")
```

Product

```python
product_table = df[['product_id', 'name',"price"]].drop_duplicates().sort_values('product_id')
print(product_table.to_string(index=False))
```

Total Quantity Sold by Product

```python
quantity_by_product = df.groupby('product_id')['quantity'].sum().sort_values(ascending=False)

quantity_by_product.plot(kind='bar', figsize=(10,5))
plt.title('Total Quantity Sold by Product')
plt.ylabel('Quantity')
plt.xlabel('Product ID')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
print("Total Quantity Sold by Product:")
print(quantity_by_product.to_frame().reset_index().to_string(index=False))
print(f"Total: {quantity_by_product.sum()}")
```

```python
# Print top 5 most sold products
print("Top 5 Most Sold Products:")
print(quantity_by_product.head(5).to_frame().reset_index().to_string(index=False))
```

```python
# Print bottom 5 least sold products
print("Bottom 5 Least Sold Products:")
print(quantity_by_product.tail(5).to_frame().reset_index().to_string(index=False))
```

Total Sales by Product

```python
sales_by_product = df.groupby('product_id')['total'].sum().sort_values(ascending=False)

sales_by_product.plot(kind='bar', figsize=(10,5))
plt.title('Total Sales by Product')
plt.ylabel('Sales (₱)')
plt.xlabel('Product ID')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
print("Total Sales by Product:")
print(sales_by_product.to_frame().reset_index().to_string(index=False))
```

```python
# Print top 5 most income in products
print("Top 5 Most Income in Products:")
print(sales_by_product.head(5).to_frame().reset_index().to_string(index=False))
```

```python
# Print top 5 least income in products
print("Top 5 Least Income in Products:")
print(sales_by_product.tail(5).to_frame().reset_index().to_string(index=False))
```
